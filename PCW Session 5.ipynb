```python
# YOUR CODE HERE
#selection_sort
def selection_sort(A):
    counter = 0
    for i in range(len(A)-1):
        minIndex = i
        counter +=1
        for j in range(i+1,len(A)):
            counter+=1
            if A[j]<A[minIndex]:
                minIndex=j
                A[i],A[minIndex] = A[minIndex],A[i]
    return A,counter
arr = [46,36,29,10,5,5,3,1]
print(selection_sort(arr))

#bubble_sort
def bubble_sort(A):
    n= len(A)
    counter = 0
    for i in range(n-1,0,-1):
        counter +=1
        for j in range(i):
            if A[j]<A[j+1]:
                A[j],A[j+1] =A[j+1],A[j]
                counter+=1
    return A,counter
arr = [3,4,5,6,6]
print(bubble_sort(arr)) 

#insertion_sort:
def insertion_sort(A):
    counter = 0
    for i in range(1,len(A)):
        key = A[i]
        j = i-1
        counter +=1
        while j>=0 and A[j]>key:
            A[j+1] = A[j]
            j-=1
            counter+=1
        A[j+1]=key
    return A,counter
arr = [3, 5, 5, 10, 29]
print(insertion_sort(arr))      

raise NotImplementedError()
```
